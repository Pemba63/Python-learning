```python
### Task 5

1)Write about different types of operatorss in python

1. Arithmetic Operators
- Used for math.

+   # addition
-   # subtraction
*   # multiplication
/   # division (always returns float)
%   # modulus (remainder)
**  # exponentiation (power)
Example:

a = 10
b = 3

print(a + b)   # 13
print(a % b)   # 1
print(a ** b)  # 1000

2. Assignment Operators
- Used to assign or update values.

=    # assign
+=   # add and assign
-=   # subtract and assign
*=   # multiply and assign
/=   # divide and assign
Example:

x = 5
x += 2   # same as x = x + 2
print(x) # 7

3. Comparison (Relational) Operators

-Used to compare values â†’ result is True or False.

==   # equal to
!=   # not equal to
>    # greater than
<    # less than
>=   # greater than or equal to
<=   # less than or equal to
Example:

a = 10
b = 20

print(a < b)   # True

4. Logical Operators
- Used to combine conditions.

and   # both conditions must be True
or    # at least one condition True
not   # reverses the result
Example:

age = 18
print(age >= 18 and age < 60)  # True

5. Bitwise Operators
- Work on binary (bits). Mostly used in low-level programming.

&   # AND
|   # OR
^   # XOR
~   # NOT
<<  # left shift
>>  # right shift
Example:

a = 5   # 101
b = 3   # 011

print(a & b)  # 1
6. Membership Operators
 - Check if a value exists in a sequence.

in
not in
Example:

fruits = ["apple", "banana", "mango"]

print("apple" in fruits)      # True
print("grape" not in fruits)  # True

7. Identity Operators
Check if two variables point to the same object in memory (not just equal).

is
is not
Example:

a = [1, 2, 3]
b = a

print(a is b)      # True
print(a == b)      # True (values are same)

8. Ternary (Conditional) Operator
One-line if-else.

result = "Pass" if marks >= 40 else "Fail"


```
