1. Arithmetic Operators
- Used for math.

+   # addition
-   # subtraction
*   # multiplication
/   # division (always returns float)
%   # modulus (remainder)
**  # exponentiation (power)
Example:

a = 10
b = 3

print(a + b)   # 13
print(a % b)   # 1
print(a ** b)  # 1000

2. Assignment Operators
- Used to assign or update values.

=    # assign
+=   # add and assign
-=   # subtract and assign
*=   # multiply and assign
/=   # divide and assign
Example:

x = 5
x += 2   # same as x = x + 2
print(x) # 7

3. Comparison (Relational) Operators

-Used to compare values â†’ result is True or False.

==   # equal to
!=   # not equal to
>    # greater than
<    # less than
>=   # greater than or equal to
<=   # less than or equal to
Example:

a = 10
b = 20

print(a < b)   # True

4. Logical Operators
- Used to combine conditions.

and   # both conditions must be True
or    # at least one condition True
not   # reverses the result
Example:

age = 18
print(age >= 18 and age < 60)  # True

5. Bitwise Operators
- Work on binary (bits). Mostly used in low-level programming.

&   # AND
|   # OR
^   # XOR
~   # NOT
<<  # left shift
>>  # right shift
Example:

a = 5   # 101
b = 3   # 011

print(a & b)  # 1
6. Membership Operators
 - Check if a value exists in a sequence.

in
not in
Example:

fruits = ["apple", "banana", "mango"]

print("apple" in fruits)      # True
print("grape" not in fruits)  # True

7. Identity Operators
Check if two variables point to the same object in memory (not just equal).

is
is not
Example:

a = [1, 2, 3]
b = a

print(a is b)      # True
print(a == b)      # True (values are same)

8. Ternary (Conditional) Operator
One-line if-else.

result = "Pass" if marks >= 40 else "Fail"

- Function in python are declared once and used multiple times

- Syntax :
     ```
    
    1. Function definition and declaration:

     def function_name(parameters):

        function_name(arguments)


    ```


````python
## function without parameter

def greeting():
    name = input("Enter your name:")
    print(f"Welcome to our profession class {name}")

greeting()
````

````python
## function with parameter 

def area(r):
    print(f"The area of circle with radius {r} is {3.14*r**2}")

print("We are calculating area of circle:")
radius = float(input("Enter the radius of circle")) 

area(radius)
````

````python
def area(l):
    print(f"The area of square with length {l} is {l**2}")

def perimeter(l,b):
    print(f"The perimeter of square with length {l} and breadth {b} is {l*b}")


length = float(input("Enter the length of square"))
breadth = float(input("Enter the breadth of square"))  
area(length)
perimeter(length,breadth)
````

## Task 1

Calculate the area of square, rectangle ,triangle  and circle using function . Use if -else ladder


Enter your value between 1-5
1. square=>area_square()
2. rectangle=>area_rectangle()
3. triangle=>area_triangle()
4. circle=>area_circle()
5. exit


````python
def area_square(l):
    
    return l**2
def area_rectangle(l, b):
   
    return l*b
def area_triangle(b, h):
   
    return 1/2*b*h
def area_circle(r):
   
    return 3.14*r**2


 # Menu
while(True):
    print("Enter your value between 1-5")
    print("1. square")
    print("2. rectangle")
    print("3. triangle")   
    print("4. circle")
    print("5. exit")
    choice = int(input("Enter your choice:"))
    # if-else ladder
    if(choice == 1):
        length = float(input("Enter the length of square"))
        print("Area of square =", area_square(length))
    elif(choice == 2):
         length = float(input("Enter the length of square"))
         breadth = float(input("Enter the breadth of rectangle"))
         print("Area of rectangle =", area_rectangle(length,breadth))
    elif(choice == 3):
         breadth = float(input("Enter the breadth of rectangle"))
         height = float(input("Enter the height of triangle"))
         print("Area of triangle =", area_triangle(breadth,height))
    elif(choice == 4):
         radius = float(input("Enter the radius of circle"))
         print("Area of circle =", area_circle(radius))
    elif(choice == 5):
        print("Exiting the program...")
        break
    else:
        print("Invalid choice")
````
